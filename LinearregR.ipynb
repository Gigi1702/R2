Linear regression in R

```r
data = read.csv('regrex1.csv') 
```

```r
plot(data$x, data$y)
```

```r
lmData= lm(y~x, data = data)
```

```r
lmData
```

```r
summary(lmData)
```

```r
pred <- predict(lmData, data)
```

```r
plot(data$x, data$y)
lines(data$x, pred)
title('Linear regression Model R')
```
